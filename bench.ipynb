```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bench.dat', comment='#', sep='\t', header=None, names=['task', 'name', 'format', 't'])
```

```python
df_ping = df[df.name.str.startswith('ping')]
df_stor = df[~df.name.str.startswith('ping')]
```

```python
def plot(df, task, title=None):
    sns.boxplot(data=df[df.task == task], y='t', x='name', hue='format', showfliers=False)
    plt.yscale('log')
    plt.ylabel('$t$ [s]')
    plt.xlabel('payload')
    if title:
        plt.title(title)

    num_categories = len(df[df.task == task]['name'].unique())
    for i in range(num_categories - 1):
        plt.axvline(i + 0.5, color='grey', linestyle='--', alpha=0.5)

    if title:
        plt.savefig(f'{title.replace(' ', '_')}.png')
```

```python
plot(df_ping, 'ser', 'cluster: Serialization')
```

```python
plot(df_ping, 'de', 'cluster: Deserialization')
```

```python
plot(df_stor, 'ser', 'storage: Serialization')
```

```python
plot(df_stor, 'de', 'storage: Deserialization')
```
